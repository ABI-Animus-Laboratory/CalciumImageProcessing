```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
#Example of find peaks method from scipy.signal

calcium_1 = np.load('/hpc/mzhu843/image-processing/minian/results/experiment5/temporal update/first update C 1.npy')
example = calcium_1[1]
peaks, _ = find_peaks(example, height=0.1)
plt.figure(figsize=(18, 5)) 
plt.plot(example, linewidth=0.8)
plt.scatter(peaks, example[peaks], s=6, color='red')
```

```python

update = 'first'

for value in sparseness_values:
    print("Sparseness Value: " + str(value))
    total_ratio = 0
    calcium = np.load(f'/hpc/mzhu843/image-processing/minian/results/experiment5/temporal update/{update} update C {value}.npy') 
    spikes = np.load(f'/hpc/mzhu843/image-processing/minian/results/experiment5/temporal update/{update} update S {value}.npy') 
    plt.show()
    for i in range(np.size(calcium, axis=0)):
        num_spikes = np.size(np.where(spikes[i])[0], axis=0)
        peaks, _ = find_peaks(calcium[i])
        total_ratio += (np.size(peaks, axis=0) / num_spikes)
        
        #For visualisation and testing only
        if i == 4:
            plt.plot(spikes[i])
            plt.scatter(peaks, calcium[i][peaks] / (np.max(calcium[i][peaks]) / np.max(spikes[i])), s=5, c='red')
            plt.show()
            plt.close()
            plt.plot(calcium[i])
    print(f'Average peak-spike ratio: {total_ratio / np.size(calcium, axis=0)}')
    print()

```

```python
sparseness_values = [1, 2, 3, 4, 5]
update = 'first'

def min_max(an_array):
    return (an_array - np.min(an_array)) / (np.max(an_array) - (np.min(an_array)))

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12), sharey=False)
axes = axes.flatten()
for i, ax in enumerate(axes):
    if i < len(sparseness_values):
        calcium = np.load(f'/hpc/mzhu843/image-processing/minian/results/experiment5/temporal update/{update} update C {sparseness_values[i]}.npy') 
        spikes = np.load(f'/hpc/mzhu843/image-processing/minian/results/experiment5/temporal update/{update} update S {sparseness_values[i]}.npy') 
        raw = np.load(f'/hpc/mzhu843/image-processing/minian/results/experiment5/temporal update/{update} update YrA {sparseness_values[i]}.npy') 
        if i == 0:
            ax.plot(min_max(raw[82]), linewidth=0.2, label='Raw')
            ax.plot(min_max(calcium[83]), label='Fitted Calcium', linewidth=0.8)
            ax.plot(min_max(spikes[83]), label='Fitted Spikes', linewidth=1.2)
            ax.set_title(f'Sparseness: {sparseness_values[i]}', fontsize=14)
            ax.legend()
        if i == 1:
            ax.plot(min_max(raw[81]), linewidth=0.2, label='Raw')
            ax.plot(min_max(calcium[82]), label='Fitted Calcium', linewidth=0.8)
            ax.plot(min_max(spikes[82]), label='Fitted Spikes', linewidth=1.2)
            ax.set_title(f'Sparseness: {sparseness_values[i]}', fontsize=14)
        if i == 2:
            ax.plot(min_max(raw[75]), linewidth=0.2, label='Raw')
            ax.plot(min_max(calcium[76]), label='Fitted Calcium', linewidth=0.8)
            ax.plot(min_max(spikes[76]), label='Fitted Spikes', linewidth=1.2)
            ax.set_title(f'Sparseness: {sparseness_values[i]}', fontsize=14)
        if i == 3:
            ax.plot(min_max(raw[67]), linewidth=0.2, label='Raw')
            ax.plot(min_max(calcium[68]), label='Fitted Calcium', linewidth=0.8)
            ax.plot(min_max(spikes[68]), label='Fitted Spikes', linewidth=1.2)
            ax.set_title(f'Sparseness: {sparseness_values[i]}', fontsize=14)
        if i == 4:
            ax.plot(min_max(raw[60]), linewidth=0.2, label='Raw')
            ax.plot(min_max(calcium[61]), label='Fitted Calcium', linewidth=0.8)
            ax.plot(min_max(spikes[61]), label='Fitted Spikes', linewidth=1.2)
            ax.set_title(f'Sparseness: {sparseness_values[i]}', fontsize=14)
            ax.set_xlabel('#Frame')
    else:
        ax.axis('off')

fig.tight_layout()
fig.subplots_adjust(top=0.90)

plt.show()
```
